## Color column filling with random rgba

```python

def random_rgba():
    return (np.random.randint(0, 256),  # Red
            np.random.randint(0, 256),  # Green
            np.random.randint(0, 256))  # Blue

def categorize_color(rgb):
    r, g, b = rgb
    max_color = max(r, g, b)
    min_color = min(r, g, b)

    if r == g == b:
        if max_color < 128:
            return 'Black'
        else:
            return 'White'
    elif max_color == r:
        if g > b:
            return 'Red-Orange' if r > 200 and g > 100 else 'Red'
        else:
            return 'Red-Purple' if r > 200 and b > 100 else 'Red'
    elif max_color == g:
        if r > b:
            return 'Yellow-Green' if g > 200 and r > 100 else 'Green'
        else:
            return 'Blue-Green' if g > 200 and b > 100 else 'Green'
    elif max_color == b:
        if r > g:
            return 'Blue-Purple' if b > 200 and r > 100 else 'Blue'
        else:
            return 'Blue'


np.random.seed(123)
df['Color'] = [random_rgba() for _ in range(len(df))]
df['Color_Name'] = df['Color'].apply(categorize_color)


```

```python
Category_ID = {"0":"Upper","1":"Lower", "2":"Over","3":"Shoes"}
category_Type_0 = {"0":"Tshirt","1":"croptop","2":"bodysuit","3":"hoodie","4":"blouse","5":"buttonshirt","6":"sweatshirt","7":"dress"}
category_Type_1 = {"0":"pants","1": "short","2":"skirt","3":"legging"}
category_Type_2 = {"0":"coat","1": "cardigan","2":"blazer","3":"jacket"}
category_Type_3 = {"0":"sandals","1": "sneakers","2":"highheels","3":"flats","4":"boots"}


def assign_occasion(row):
    category_to_occasion = {
        'Tshirt': '1001',  # Casual and Sport
        'blouse': '1100',  # Formal and Casual #casual , formal,party,sport
        'croptop': '1011',
        'bodysuit': "0010",
        "hoodie": "1001",
        "buttonshirt":"0100",
        "sweatshirt":"1001",
        "dress":"0110",
        "pants":"1111",
        "short" :"1011",
        "skirt":"0110",
        "legging":"1001",
        "coat":"1110",
        "cardigan":"1000",
        "blazer":"0110",
        "jacket":"1111",
        "sandals":"1010",
        "sneakers":"1001",
        "highheels":"0110",
        "flats":"1110",
        "boots":"1110"}



        # Extend this dictionary for other categories with their corresponding occasion codes

    # Return the occasion code if the category is in the dictionary, else return a default value
    return category_to_occasion.get(row['Category'], '0000')
df['Occasion'] = df.apply(assign_occasion, axis=1)
# for each row  (axis = 1 means for each row across the columns) get the category of that row and assign the corresponding value according to the dictionary)
```

```python
def is_nearly_white_or_dark(color):
    r, g, b = color
    # Define thresholds
    dark_threshold = 50  # Colors are considered dark if below this threshold
    white_threshold = 205  # Colors are considered nearly white if above this threshold
    # Check if color is nearly white or dark
    is_dark = r < dark_threshold and g < dark_threshold and b < dark_threshold
    is_nearly_white = r > white_threshold and g > white_threshold and b > white_threshold
    return is_dark or is_nearly_white

def adjust_occasion_for_color(row):
    # Extract the 'Occasion' and 'Color'
    occasion = row['Occasion']
    color = row['Color']
    # Check if second character is '1' (formal)
    if occasion[1] == '1':
        # If color does not meet the criteria, set the second character to '0'
        if not is_nearly_white_or_dark(color):
            occasion = occasion[0] + '0' + occasion[2:]
    return occasion

# Apply the function to each row in the DataFrame
df['Occasion'] = df.apply(adjust_occasion_for_color, axis=1)
```

```python
#converting RGB to HSV
def rgb_to_hsv(r, g, b,precision=2):
    r, g, b = r / 255.0, g / 255.0, b / 255.0
    max_rgb = max(r, g, b)
    min_rgb = min(r, g, b)
    delta = max_rgb - min_rgb

    if delta == 0:
        h = 0
    elif max_rgb == r:
        h = (60 * ((g - b) / delta) + 360) % 360
    elif max_rgb == g:
        h = (60 * ((b - r) / delta) + 120) % 360
    else:
        h = (60 * ((r - g) / delta) + 240) % 360

    if max_rgb == 0:
        s = 0
    else:
        s = delta / max_rgb

    v = max_rgb

    return round(h), round(s, precision), round(v, precision)

df['HSV'] = df['Color'].apply(lambda x: rgb_to_hsv(*x))
df
```

```python
def is_complementary(hsv1, hsv2):
    # Calculate the absolute difference in hue
    hue_diff = abs(hsv1[0] - hsv2[0])
    # Check if the hues are 180 degrees apart (with a small margin of error)
    return abs(hue_diff - 180) < 10

def is_harmonic(hsv1, hsv2):
    # Calculate the absolute difference in hue
    hue_diff = abs(hsv1[0] - hsv2[0])
    # Check if the hues are within 30 degrees of each other
    return hue_diff < 30 or hue_diff > 330  # Considering the circular nature of the color wheel


def is_analogous(hsv1, hsv2):
    # Calculate the absolute difference in hue
    hue_diff = abs(hsv1[0] - hsv2[0])

    # Adjust for the circular nature of the color wheel
    if hue_diff > 180:
        hue_diff = 360 - hue_diff

    # Check if the hues are within 30 degrees of each other
    return hue_diff <= 30
```

```python
def FFind_unique_outfit(df, occasion=None, include_over=False):
    df_occ = df[df[occasion] == 1] if occasion else df
    if df_occ[df_occ['ID'] == 0].empty:
        return "No upper items available for selection."

    upper_items = df_occ[df_occ['ID'] == 0]
    lower_items = df_occ[df_occ['ID'] == 1]
    shoes = df_occ[df_occ['ID'] == 3]
    overs = df_occ[df_occ['ID'] == 2] if include_over else pd.DataFrame()

    selected_uppers = upper_items if len(upper_items) <= 3 else upper_items.sample(n=3, replace=False)

    color_relations = [is_harmonic, is_complementary, is_analogous]
    relation_names = ['Harmonic', 'Complementary', 'Analogous']
    outfits = []

    for _, upper in selected_uppers.iterrows():
        outfit_attempts = []
        attempts = 0

        while len(outfit_attempts) < 3 and attempts < 6:
           # Check for uniqueness

            relation = color_relations[attempts % len(color_relations)]
            relation_name = relation_names[attempts % len(color_relations)]
            upper_hsv = upper['HSV']
            is_dress = upper['Category'].lower() == 'dress'

            shoe_candidates = shoes[shoes['HSV'].apply(lambda x: relation(x, upper_hsv))]
            lower_candidates = lower_items[lower_items['HSV'].apply(lambda x: relation(x, upper_hsv))] if not is_dress else pd.DataFrame()
            over_candidates = overs[overs['HSV'].apply(lambda x: relation(x, upper_hsv))] if include_over else pd.DataFrame()

            if not shoe_candidates.empty:
                shoe = shoe_candidates.sample(n=1)
            else:
                attempts += 1
                continue

            if is_dress or not lower_candidates.empty:
                lower = lower_candidates.sample(n=1) if not is_dress else pd.DataFrame()
            else:
                attempts += 1
                continue

            if include_over and not over_candidates.empty:
                over = over_candidates.sample(n=1)
            else:
                over = pd.DataFrame()

            # Construct a tentative outfit
            tentative_outfit = [("Upper", upper['Category'], upper['Color'],relation_name)]
            if not shoe.empty:
                tentative_outfit.append(("Shoes", shoe.iloc[0]['Category'], shoe.iloc[0]['Color'],relation_name))
            if not lower.empty:
                tentative_outfit.append(("Lower", lower.iloc[0]['Category'], lower.iloc[0]['Color'],relation_name))
            if not over.empty:
                tentative_outfit.append(("Over", over.iloc[0]['Category'], over.iloc[0]['Color'],relation_name))

            outfit_tuple = tuple([item for item in tentative_outfit])



            # Check for uniqueness
            if outfit_tuple not in outfit_attempts:
                outfit_attempts.append(outfit_tuple)
                outfits.append(tentative_outfit)  # Add to final outfits list as the formatted structure

            attempts += 1

    # Format the output to include category, color, and relation type for each outfit item
    formatted_outfits = []
    for outfit in outfits:
        formatted_outfit = {}
        for item in outfit:
            formatted_outfit[item[0]] = {'Category': item[1], 'Color': item[2],'relation': item[3]}
        formatted_outfits.append(formatted_outfit)

    return formatted_outfits
```
