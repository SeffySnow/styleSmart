```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
```

```python
# Load the image
image = cv2.imread('colorDetectionTests\pant3.jpg')

# Convert the image to HSV color space
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Flatten the image into a 2D array of pixels
pixels = hsv_image.reshape((-1, 3))

# Define the number of clusters (colors) you want to detect
num_clusters = 2 # Adjust this based on the number of colors you expect

# Apply K-means clustering
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(pixels)

# Get the representative colors (centroids) of each cluster
colors = kmeans.cluster_centers_

# Convert the colors from float to integer
colors = colors.astype(np.uint8)

# Print the representative colors
for i, color in enumerate(colors):
    print("Color {} (HSV): {}".format(i+1, color))

# Optionally, convert the colors to BGR and display them
# colors_bgr = colors
colors_bgr = cv2.cvtColor(colors.reshape(1, -1, 3), cv2.COLOR_HSV2BGR)
# colors_bgr = cv2.cvtColor(colors.reshape(1, -1, 3), cv2.COLOR_BGR2LAB)
for i, color_bgr in enumerate(colors_bgr[0]):
    print("Color {} (BGR): {}".format(i+1, color_bgr))
    
print(colors[0])

# Display the representative colors (optional)
for i, color_bgr in enumerate(colors_bgr[0]):
    cv2.imshow('Color {}'.format(i+1), np.full((100, 100, 3), color_bgr, dtype=np.uint8))
    

# cv2.imshow('Color {}'.format(i+1), np.full((100, 100, 3), enumerate(colors_bgr[0])[0], dtype=np.uint8))

cv2.imshow('Color {}'.format(i+1), np.full((100, 100, 3), color_bgr, dtype=np.uint8))

cv2.imshow('Clothes Image', image)
    
cv2.waitKey(0)
cv2.destroyAllWindows()
```
